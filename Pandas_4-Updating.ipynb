# Apply Method

```python
df1["NAME"]
```

```python
df1["NAME"].apply(len)         # get length of the NAME
```

```python
def update_name(name):
    return name.upper()
```

```python
df1["NAME"] = df1["NAME"].apply(update_name)
```

```python
df1
```

Lambda function

```python
df1["NAME"] = df1["NAME"].apply(lambda x: x.lower())
```

```python
df1
```

```python
df1["Maths 1"] = df1["Maths 1"].apply(lambda x: x+1)   # increasing marks for Maths 1 by 1 value  (Calculation)
```

```python
df1["NAME"].apply(len)         # get length of the NAME
```
