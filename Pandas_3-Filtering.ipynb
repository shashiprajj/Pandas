# Pandas Filtering

```python
import pandas as pd
```

```python
df = pd.read_csv("data/survey_results_public.csv")
```

```python
df1 = pd.read_csv("csv_file.csv")
```

```python
students = {
    "Name" : ["Shashi", "Meet", "Anshuman", "Prashant", "Vikhil"],
    "Roll-no" : [46, 64, 65, 67, 50], 
    "email" : ["shashiprajapati38@gmail.com", "meet@gmail.com", "anshuman@gmail.com", "prashant@gmail.com", "vikhil@gmail.com"]
}
```

```python
df2 = pd.DataFrame(students)
df2
```

```python
df2.set_index("Roll-no", inplace=True)   # setting roll no as ndex as it is unique for everyone
```

```python
df2
```

```python
df2.sort_index(ascending=True, inplace=True)   # sorting roll-no i.e. index in ascending order
df2
```

```python
df2["Name"] == "Shashi"  # this just returns true and false for those particular rows
```

```python
df2[df2["Name"] == "Shashi"]   # to return a dataframe of those filtered data
```

Storing in a Variable

```python
filt = (df2["Name"] == "Shashi")
```

```python
df2[filt]
```

```python
df2.loc[filt, "email"]  #to return particular columns from those filtered rows
```

```python
filt2 =  (df2["Name"] == "Shashi") & (df2["email"] == "shashiprajapati38@gmail.com")
df2[filt2]
```

```python
filt3 =  (df2["Name"] == "Shashi") | (df2["email"] == "meet@gmail.com")
df2[filt3]
```

```python
filt4 =  (df1["Maths 1"] >= 60) & (df1["Maths 2"] >= 60) 
df1[filt4]
```

```python
filt5 =  (df1["Maths 1"] >= 60) & (df1["Maths 2"] >= 60) & (df1["Maths 3"] >= 60) 
df1[filt5]
```

```python
filt6 =  (df1["Maths 1"] >= 80) | (df1["Maths 2"] >= 80)  
df1[filt6]
```

```python
filt6 =  (df1["Maths 1"] >= 80) | (df1["Maths 2"] >= 80)  
df1.loc[filt6, ["Name", "Roll no", "Batch"]]
```

```python
filt6 =  (df1["Maths 1"] >= 80) | (df1["Maths 2"] >= 80)  
df1.loc[-filt6, ["Name", "Roll no", "Batch"]]                     # inverse of filt6
```
