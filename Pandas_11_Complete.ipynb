# REGEX OPERATIONS

```python
df.loc[df["Type 1"].str.contains('Fire', regex = True)]
```

```python
df.loc[df["Type 1"].str.contains('Fire|Grass', regex = True)]
```

```python
df.loc[df["Name"].str.contains('Pi[a-z]*', regex = True)]       # Name containing Pi will be printed
```

```python
df.loc[df["Name"].str.contains('^Pi[a-z]*', regex = True)]   # Name starts from Pi will be printed
```

```python
df.loc[df["Name"].str.contains('^Pi[a-z]*', regex = True)]   # Name starts from Pi will be printed
```

Change Fire to Flamer from Type 1 column

```python
flame = (df["Type 1"] == "Fire")
```

```python
df.loc[flame, "Type 1"] = "Flamer"
```

```python
df.sort_index()
```

OR Another Method

```python
df.loc[df["Type 1"] == "Grass", "Type 1"] = "Grasses"
```

```python
df.head()
```

```python
df.loc[df["Type 1"] == "Flamer", "Legendary"] = True  # All flamer pokemons sets to True legendary
```
