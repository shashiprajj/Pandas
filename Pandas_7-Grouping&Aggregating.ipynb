# Solving Problem

```python
country_respondents = df["Country"].value_counts()
country_respondents
```

```python
country_uses_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
country_uses_python
```

```python
python_df = pd.concat([country_uses_python, country_respondents], axis="columns", sort=False)
```

```python
python_df.sort_values(by = ["Country"],ascending=False)
```

```python
python_df.rename(columns={"LanguageWorkedWith":"No knows Python", "Country":"No of Respondents"}, inplace=True)
```

```python
python_df.sort_values(by = ["No knows Python", "No of Respondents"],ascending=False)
```

from Unites states, 10083 people responded that they knows python out of 20949

from India, 3105 people responded that they knows python out of 9061

```python
python_df["perct_people"] = (python_df["No knows Python"]/ python_df["No of Respondents"]) * 100
```

```python
python_df.sort_values(by = ["perct_people"],ascending=False)
```

```python
python_df.sort_values(by = ["No knows Python", "No of Respondents"], ascending=False, inplace=True)
```
